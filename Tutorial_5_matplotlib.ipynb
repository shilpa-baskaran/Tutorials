## Ploting using Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2]
y = [10,30,60]
plt.plot(x,y)
```

```python
plt.plot(x,y);
```

```python
df = pd.DataFrame({'Home':[1,1,2,2,3,3],
                 'price':[1000,1100,700,750,200,100]})
df
```

```python
plt.scatter(df['Home'],df['price'])
```

```python
plt.plot(df['Home'],df['price'],color='red')
plt.title('House price')
plt.xlabel('Category')
plt.ylabel('Price')
```

```python
plt.plot(df['Home'],df['price'],color='red',marker='o',markersize=5,linestyle='--')
plt.xlim(1,3)
plt.ylim(100,1100)
plt.title('House price')
plt.xlabel('Category')
plt.ylabel('Price')
```
