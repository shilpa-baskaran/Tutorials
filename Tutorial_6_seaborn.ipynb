## Visualization using seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\gopin\Desktop\Tutorial\dataset\heart.csv")
```

```python
df.head()
```

## Distplot

```python
sns.distplot(df['age'])
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['age'])
```

```python
sns.distplot(df['age'],kde=False)
```

```python
sns.distplot(df['age'],kde=False,bins=40,color='red')
```

## Count plot

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='target',data=df)
```

```python
sns.countplot(x="cp",data=df)
```

```python
sns.countplot(x='cp',data=df,hue='sex')
```

## Box plot

```python
# to identify outliers
# In this the blue is female and orange is male, where most of the female is between 50 - 65 and male between 48 to 60
sns.boxplot(x = 'sex',y='age',data=df)
```

```python
sns.boxplot(x='thal',y='chol',data=df,hue='sex')
```

```python
# scaltter plot  -  x-nonlinera,y- nonlinear
# count plot - x - catagorize data
# boxplot - x - catagorize data, y - non linear
# distplot - x - nonlinear or catgorize
```

```python
sns.scatterplot(x='age',y='chol',data=df)
```
